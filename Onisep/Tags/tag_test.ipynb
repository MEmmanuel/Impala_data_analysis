```python
import os
import json
import numpy as np
import math
import pandas as pd
```

```python
# Chemins utiles
dir_path = os.getcwd()
# Tags file
tags_path = os.path.join(dir_path, 'tags_assoc_10oct.xlsx')
# All_tags file
all_tags_path = os.path.join(dir_path, 'tags_assoc_10oct(1).xlsx')
```

```python
metiers_tags = pd.read_excel(tags_path, sheetname=1, header=0)
metiers_tags.head()
all_tags = pd.read_excel(all_tags_path, sheetname=1, header=0,)
```

```python
jobs_list = np.unique(metiers_tags["job"]).tolist()
jobs_list[0:10]
```

```python
tags_sample = np.unique(metiers_tags["tag"]).tolist()
print("Nombre de tags dans le sample", len(tags_sample))
tags = np.unique(all_tags["tag"]).tolist()
print("Nombre total de tags", len(tags))
```

```python
df_mt = {}
tags_list = []
for job in jobs_list:
    df_mt[job] = []

for row in metiers_tags.itertuples():
    job = row[1]
    tag = row[2]
    if row[4]==0:
        df_mt[job].append(tag)
        if tag not in tags_list:
            tags_list.append(tag)
```

```python
tags_list
```

```python
df_score = pd.DataFrame(columns = ['métier','score'])

j = 0
for job in df_mt:
    df_score.loc[j,'métier'] = job
    df_score.loc[j,'score'] = 0
    for tag in tags:
        if tag in df_mt[job]:
            df_score.loc[j,'score'] = df_score.loc[j,'score']+1
    j+=1

df_score.head()
```

```python
df_score.sort_values(by="score",ascending=False)[0:10]
```
