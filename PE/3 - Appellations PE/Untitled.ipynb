```python
# Ce script permet de trier les appellations de Pôle Emploi

import os
import json
```

```python
dir_path = os.getcwd()
fiches_metiers = os.path.join(dir_path, 'fiches_metiers')

data_dir = os.path.join(dir_path, 'data.json')

list_metiers = os.listdir(fiches_metiers)
```

```python

jobs_appellations = {}
for file in list_metiers:
    with open(fiches_metiers + '/' + file, 'r') as f:
        job = json.load(f)
    jobs_appellations[job["code_rome"]] = [job["intitule"]]
    for item in job["appelations"]:
        jobs_appellations[job["code_rome"]].append(item["libelle"])
```

```python
# Copie du dictionnaire au format JSON
with open(data_dir, 'w') as f:
    json.dump(jobs_appellations, f, indent=4)
```
